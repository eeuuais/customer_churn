```python
# 전체 데이터를 불러오기 (메모리가 걱정된다면 필수 컬럼만!)
trans_all = pd.read_csv('./raw/transactions_train.csv', usecols=['t_dat', 'customer_id'])
trans_all['t_dat'] = pd.to_datetime(trans_all['t_dat'])

# 1. 기준일 설정
max_date = trans_all['t_dat'].max()
cut_off = max_date - pd.Timedelta(days=7)

# 2. 마지막 주 구매자(유지)와 그 전 구매자(대상) 구분
target_week_users = set(trans_all[trans_all['t_dat'] > cut_off]['customer_id'].unique())
all_users = trans_all[trans_all['t_dat'] <= cut_off]['customer_id'].unique()

# 3. 라벨링: 마지막 주에 안 나타나면 이탈(1)
df_label = pd.DataFrame({'customer_id': all_users})
df_label['churn'] = df_label['customer_id'].isin(target_week_users).astype(int).map({0: 1, 1: 0})

print(df_label['churn'].value_counts())
```

```python
# 이탈의 기준 잡기

# 유저별 구매 간격 계산을 위해 데이터 정렬
trans_sorted = trans_all.sort_values(['customer_id', 't_dat'])

# 유저별로 이전 구매일과의 차이(diff) 계산
trans_sorted['diff'] = trans_sorted.groupby('customer_id')['t_dat'].diff().dt.days

# 각 유저가 보여준 '가장 길었던 구매 공백' 추출
user_max_diff = trans_sorted.groupby('customer_id')['diff'].max().dropna()

# 전체 유저의 최대 공백 중 상위 95% 지점을 이탈 기준으로 채택
threshold_days = user_max_diff.quantile(0.95)

print(f"데이터 기반 권장 이탈 기준일: 약 {int(threshold_days)}일")
```

75% 지점의 수치가 만약 180일(6개월) 내외로 나온다면, "우리 고객의 75%는 이 기간 안에 돌아오므로, 이 이상 소식이 없으면 이탈로 간주하겠다"는 아주 훌륭한 근거가 됩니다.

```python
# 1. 전체 거래 데이터 로드 (필수 컬럼만)
trans_all = pd.read_csv('./raw/transactions_train.csv', usecols=['t_dat', 'customer_id'])
trans_all['t_dat'] = pd.to_datetime(trans_all['t_dat'])

# 2. 유저별 최대 재구매 간격 계산 (이미 계산하셨던 로직 요약)
trans_sorted = trans_all.sort_values(['customer_id', 't_dat'])
trans_sorted['diff'] = trans_sorted.groupby('customer_id')['t_dat'].diff().dt.days
user_max_diff = trans_sorted.groupby('customer_id')['diff'].max().dropna()

# 3. 이탈 기준일(Threshold) 확정: 상위 75% 지점
q75_days = int(user_max_diff.quantile(0.75))
print(f"채택된 이탈 기준일: {q75_days}일")

# 4. 라벨링 설계
# 기준점(Cut-off): 전체 데이터 종료일로부터 q75_days만큼 거슬러 올라간 날짜
max_date = trans_all['t_dat'].max()
cut_off_date = max_date - pd.Timedelta(days=q75_days)

print(f"데이터 종료일: {max_date.date()}")
print(f"라벨링 기준일: {cut_off_date.date()}")

# 5. 유저 분류
# A. 기준일(cut_off) 이전에 최소 한 번이라도 구매한 적이 있는 전체 유저 (분석 대상)
all_active_users = trans_all[trans_all['t_dat'] <= cut_off_date]['customer_id'].unique()

# B. 기준일(cut_off) 이후에 구매 기록이 있는 유저 (유지 유저)
stayed_users = set(trans_all[trans_all['t_dat'] > cut_off_date]['customer_id'].unique())

# 6. 최종 정답지(Label) 생성
df_label = pd.DataFrame({'customer_id': all_active_users})
df_label['churn'] = df_label['customer_id'].apply(lambda x: 0 if x in stayed_users else 1)

print("\n--- 최종 라벨링 결과 ---")
print(df_label['churn'].value_counts())
print(f"이탈률: {df_label['churn'].mean() * 100:.2f}%")
```

```python
# 1. 유저별 최초/마지막 구매일 요약 (전체 데이터 대상)
# trans_all은 't_dat'가 datetime 형식으로 변환된 전체 거래 데이터여야 합니다.
user_stats = trans_all.groupby('customer_id')['t_dat'].agg(['min', 'max']).reset_index()
user_stats.columns = ['customer_id', 'first_buy', 'last_buy']

# 2. 기준 지표 계산 (가입 기간 및 마지막 구매 후 경과일)
max_date = trans_all['t_dat'].max()
user_stats['tenure'] = (max_date - user_stats['first_buy']).dt.days
user_stats['recency'] = (max_date - user_stats['last_buy']).dt.days

# 3. [로직 적용 1] 가입한 지 30일 미만인 '초기 유저' 제외
# 최소한의 판단 근거가 없는 유저는 모델의 노이즈가 되므로 제거합니다.
df_label = user_stats[user_stats['tenure'] >= 30].copy()

# 4. [로직 적용 2] 유저별 맞춤 이탈 기준일(Threshold) 설정
# 가입 기간이 203일보다 작으면 가입 기간을, 크면 203일을 기준으로 삼음
df_label['threshold'] = df_label['tenure'].apply(lambda x: min(x, 203))

# 5. [로직 적용 3] 이탈 여부(Churn) 결정
# '마지막 구매 후 흐른 시간'이 '본인의 판단 기준일'보다 크거나 같으면 이탈(1)
df_label['churn'] = (df_label['recency'] >= df_label['threshold']).astype(int)

# 결과 확인
print(f"✅ 분석 대상 유저 수: {len(df_label):,}명 (30일 미만 제외 완료)")
print(f"✅ 이탈률 현황:\n{df_label['churn'].value_counts(normalize=True).map(lambda x: f'{x:.2%}')}")
```

```python
# 1. 필요한 컬럼 4개만 지정해서 다시 로드
# t_dat(날짜), customer_id(고객), article_id(상품), price(가격)
trans_for_rfm = pd.read_csv('./raw/transactions_train.csv', 
                            usecols=['t_dat', 'customer_id', 'article_id', 'price'])

# 2. 불러오자마자 유저별로 요약 (메모리 확보를 위해)
user_rfm = trans_for_rfm.groupby('customer_id').agg({
    'article_id': 'count',          # Frequency
    'price': ['mean', 'sum']        # Monetary
}).reset_index()

# 3. 컬럼명 정리
user_rfm.columns = ['customer_id', 'frequency', 'price_mean', 'price_sum']

# 4. 요약이 끝났으면 큰 데이터는 메모리에서 삭제
del trans_for_rfm

# 5. 기존에 만들어둔 df_label에 합치기
df_model = pd.merge(df_label, user_rfm, on='customer_id', how='left')

# 6. 확인
print("--- RFM 데이터 결합 완료 ---")
print(df_model.head())
```

```python
# 1. customers에서 분석에 필요한 컬럼들 선택 (postal_code 제외)
customers_features = customers[[
    'customer_id', 
    'FN', 
    'Active', 
    'club_member_status', 
    'fashion_news_frequency', 
    'age'
]]

# 2. 기존 df_model(RFM + Label)과 customers 정보 결합
# how='left'를 사용해 133만 명의 라벨링된 유저를 기준으로 정보를 붙입니다.
df_analysis = pd.merge(df_model, customers_features, on='customer_id', how='left')

# 3. 데이터 구성 확인
print("--- 이탈 분석용 마스터 데이터셋 구성 ---")
print(df_analysis.info())
```

```python
# 1. 중복 고객이 없는지 다시 한번 확인
print(f"중복 없는 고객 수: {df_analysis['customer_id'].nunique()}")
print(f"전체 데이터 행 수: {len(df_analysis)}")

# 2. 이탈자(1)와 유지자(0)의 평균 구매 횟수(frequency) 비교
# 데이터가 잘 만들어졌다면 0(유지)인 그룹의 구매 횟수가 훨씬 높게 나올 겁니다.
print("\n--- 이탈 여부별 평균 구매 횟수 ---")
print(df_analysis.groupby('churn')['frequency'].mean())
```

```python
# 1. article_id를 가져오기 위해 필요한 컬럼만 지정해서 다시 로드
# 전체를 다시 부르는 게 아니라, id 연결을 위해 필요한 것만 읽어옵니다.
trans_article_link = pd.read_csv('./raw/transactions_train.csv', 
                                 usecols=['customer_id', 'article_id'])

# 2. 분석 대상인 133만 명의 데이터만 남기기 (메모리 확보)
# df_analysis에 있는 고객들만 남깁니다.
temp = trans_article_link[trans_article_link['customer_id'].isin(df_analysis['customer_id'])]

# 3. articles 정보와 병합
temp = pd.merge(temp, 
                articles[['article_id', 'index_group_name', 'product_group_name']], 
                on='article_id', how='left')

# 4. 유저별 최빈값 계산 (속도가 빠른 value_counts 방식)
def get_most_frequent(x):
    if x.empty: return "NONE"
    return x.value_counts().idxmax()

print("카테고리 최빈값을 계산 중입니다... (잠시만 기다려주세요)")
user_style = temp.groupby('customer_id').agg({
    'index_group_name': get_most_frequent,
    'product_group_name': get_most_frequent
}).reset_index()

# 5. 마스터 데이터셋(df_analysis)에 최종 병합
df_analysis = pd.merge(df_analysis, user_style, on='customer_id', how='left')

# 6. 사용한 임시 객체 삭제
del trans_article_link, temp, user_style

print("--- 취향 정보 합치기 완료 ---")
print(df_analysis[['customer_id', 'index_group_name', 'product_group_name']].head())
```
